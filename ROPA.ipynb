```python
import pandas as pd
# read data and put it in a dataframe
gsheets = gc.open_by_url(SS_Link)
sheets = gsheets.worksheet('Data Infos').get_all_values()
df_infos = pd.DataFrame(sheets[1:], columns=sheets[0])

df_infos
```

```python

for i in range(0,len(df_infos)):
  z = df_infos.iloc[i, 0]
  if z == '':
    df_infos.iloc[i, 0] = df_infos.iloc[i-1, 0]

df_infos#.iloc[:, 0].unique().tolist()
```

```python
df_sample = df_infos[df_infos.iloc[:, 0] == 'System I']

df_sample = df_sample.iloc[:, [0,7,11,12]]#,14,19]]

df_sample
```

```python
df_sample.columns = ['System','Data_Category','Data_Purpose','Data_Uses']#,'','']
```

```python

df_sample

for i in range(0,len(df_sample)):
  z = df_sample.iloc[i, 1]
  if z == '':
    df_sample.iloc[i, 1] = df_sample.iloc[i-1, 1]

df_sample#.iloc[:, 0].unique().tolist()
```

```python
df_sample_gp = df_sample.groupby(['System', 'Data_Category'])['Data_Purpose'].apply('||'.join).reset_index()

df_sample_gp
```

```python
df_sample_unique = df_sample[['System','Data_Category']].drop_duplicates()

df_sample_unique
```

```python
df_merged = pd.merge(df_sample_unique, df_sample_gp, how='left', left_on=['System', 'Data_Category'], right_on=['System', 'Data_Category'] )

df_merged = pd.merge(df_merged, df_sample, how='left', left_on=['System', 'Data_Category'], right_on=['System', 'Data_Category'] )


#df_merged['System','Data_Category','Data_Purpose_y','Data_Uses']

df_merged = df_merged[['System','Data_Category','Data_Purpose_y','Data_Uses']]
df_merged
```

```python
df_merged.iloc[1,[2,3]]
```

```python
df_merged_sample = df_merged

df_merged_sample['Data_Uses'] = df_merged_sample['Data_Uses'].apply(lambda x: x.replace("\n","||"))

df_merged_sample['Data_Uses'] = df_merged_sample['Data_Uses'].apply(lambda x: x.split('||'))
df_merged_sample['Data_Purpose_y'] = df_merged_sample['Data_Purpose_y'].apply(lambda x: x.split('||'))

df_merged_sample
```

```python
def product(*args):
    if not args:
        return iter(((),)) # yield tuple()
    return (items + (item,)
            for items in product(*args[:-1]) for item in args[-1])

list(product([1,2,3], ['a','b']))
```

```python
df_merged_sample_zip = df_merged_sample

df_merged_sample_zip['zip'] = df_merged_sample_zip.apply(lambda x: list(product(x['Data_Purpose_y'], list(x['Data_Uses']))), axis=1)

df_merged_sample_zip = df_merged_sample_zip[['System','Data_Category','zip']]

df_merged_sample_zip
```

```python
df_merged_sample_zip['join_trick'] = ''
df_merged_sample_zip
```

```python
df_merged_sample_zip.iloc[1,1]
```

```python
for r in range(0, len(df_merged_sample_zip)):
  #print(df_merged_sample_zip.iloc[r,2])

  if df_merged_sample_zip.iloc[r,2] == [('', '')] or df_merged_sample_zip.iloc[r,2] == [('Same as the previous data category', 'Same as the previous data category')]:
    #print('ok')
    anc = df_merged_sample_zip.iloc[r-1,1]
    #print(anc)
    new=df_merged_sample_zip.iloc[r-1,3]

    if len(df_merged_sample_zip.iloc[r-1,3])==0:
      df_merged_sample_zip.iloc[r,3] = anc
    else:
      df_merged_sample_zip.iloc[r,3] = new



  # else:
  #   new = df_merged_sample_zip.iloc[r,1]
  #   #print(anc)
  #   df_merged_sample_zip.iloc[r,3] = new

df_merged_sample_zip
```

```python
df_merged_sample_zip[df_merged_sample_zip['join_trick']=='']
```

```python
df_merged_sample_zip_same = pd.merge(df_merged_sample_zip, df_merged_sample_zip, how='inner', left_on=['System', 'join_trick'], right_on=['System', 'Data_Category'] )

df_merged_sample_zip_same = df_merged_sample_zip_same[['System','Data_Category_x','zip_y','join_trick_x']]

df_merged_sample_zip_same.columns = ['System','Data_Category','zip','join_trick']

df_merged_sample_zip_same
```

```python
df_merged_sample_zip_fin = pd.concat([
    df_merged_sample_zip[df_merged_sample_zip['join_trick']==''],
    df_merged_sample_zip_same], ignore_index=True)

df_merged_sample_zip_fin
```

```python
df_merged_sample_zip_fin.iloc[1,2]
```

```python
df_merged_sample_zip_fin_explode = df_merged_sample_zip_fin[['System','Data_Category','zip']].explode('zip')

df_merged_sample_zip_fin_explode
```

```python
df_merged_sample_zip_fin_explode['Data_Purpose'] = df_merged_sample_zip_fin_explode['zip'].apply(lambda x: x[0] )

df_merged_sample_zip_fin_explode['Data_Uses'] = df_merged_sample_zip_fin_explode['zip'].apply(lambda x: x[1] )

df_merged_sample_zip_fin_explode
```

```python
df_final = df_merged_sample_zip_fin_explode[['System','Data_Category','Data_Purpose','Data_Uses']]
df_final
```
