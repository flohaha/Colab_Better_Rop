# Treat the explosion Data Category / Sources / Purpose

```python
df_sample_unique = df_sample[['System','Sources','Data_Category']].drop_duplicates()

df_sample_unique.head(5)
```

```python
df_merged = pd.merge(df_sample_unique, df_sample_gp, how='left', left_on=['System','Sources', 'Data_Category'], right_on=['System', 'Sources', 'Data_Category'] )

df_merged = pd.merge(df_merged, df_sample, how='left', left_on=['System','Sources', 'Data_Category'], right_on=['System','Sources', 'Data_Category'] )


#df_merged['System','Sources','Data_Category','Data_Purpose_y','Data_Uses']

df_merged = df_merged[['System','Sources','Data_Category','Data_Purpose_y','Data_Uses']]

df_merged.head()
```

```python
df_merged_sample = df_merged

df_merged_sample['Data_Uses'] = df_merged_sample['Data_Uses'].apply(lambda x: x.replace("\n","||"))

df_merged_sample['Data_Uses'] = df_merged_sample['Data_Uses'].apply(lambda x: x.split('||'))
df_merged_sample['Data_Purpose_y'] = df_merged_sample['Data_Purpose_y'].apply(lambda x: x.split('||'))

df_merged_sample.head()
```

```python
def product(*args):
    if not args:
        return iter(((),)) # yield tuple()
    return (items + (item,)
            for items in product(*args[:-1]) for item in args[-1])

list(product([1,2,3], ['a','b']))
```

# Treat the Data Category for Same As Previous

```python
df_merged_sample_zip['join_trick'] = ''
df_merged_sample_zip.head()
```

```python
for r in range(0, len(df_merged_sample_zip)):
  #print(df_merged_sample_zip.iloc[r,2])

  if df_merged_sample_zip.iloc[r,3] == [('', '')] \
    or df_merged_sample_zip.iloc[r,3] == [('Same as the previous data category', 'Same as the previous data category')] \
    or len([s for s in df_merged_sample_zip.iloc[r,3][0]  if 'Same as' in s])>0 \
    or len([s for s in df_merged_sample_zip.iloc[r,3][0]  if 'same as' in s])>0 :


    anc = df_merged_sample_zip.iloc[r-1,2]
    #print(anc)
    new=df_merged_sample_zip.iloc[r-1,4]

    if len(df_merged_sample_zip.iloc[r-1,4])==0:
      df_merged_sample_zip.iloc[r,4] = anc
    else:
      df_merged_sample_zip.iloc[r,4] = new



  # else:
  #   new = df_merged_sample_zip.iloc[r,1]
  #   #print(anc)
  #   df_merged_sample_zip.iloc[r,3] = new

df_merged_sample_zip.head()
```

```python
df_merged_sample_zip_same = pd.merge(df_merged_sample_zip, df_merged_sample_zip, how='inner',
                                     left_on=['System', 'Sources', 'join_trick'], right_on=['System','Sources', 'Data_Category'] )

df_merged_sample_zip_same = df_merged_sample_zip_same[['System','Sources','Data_Category_x','zip_y','join_trick_x']]

df_merged_sample_zip_same.columns = ['System', 'Sources','Data_Category','zip','join_trick']

df_merged_sample_zip_same.head()
```

```python
df_merged_sample_zip_fin = pd.concat([
    df_merged_sample_zip[df_merged_sample_zip['join_trick']==''],
    df_merged_sample_zip_same], ignore_index=True)

df_merged_sample_zip_fin.head()
```
